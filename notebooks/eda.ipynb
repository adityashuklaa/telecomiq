# TelecomIQ — Exploratory Data Analysis

Quick EDA over the generated telecom datasets used to train the three models.
Run `python -m data.generate_data` first (from the repo root).

```python
import sys, pathlib
sys.path.append(str(pathlib.Path.cwd().parent))  # repo root on path
import pandas as pd

DATA = pathlib.Path('../data/generated')
customers = pd.read_csv(DATA / 'customers.csv')
cdrs = pd.read_csv(DATA / 'cdrs.csv')
network = pd.read_csv(DATA / 'network.csv')
print('customers', customers.shape, '| cdrs', cdrs.shape, '| network', network.shape)
customers.head()
```

## Churn — class balance and key drivers

```python
print('Churn rate: %.1f%%' % (100 * customers['churn'].mean()))
customers.groupby('contract')['churn'].mean().round(3)
```

```python
customers.groupby(pd.cut(customers['tenure_months'], [0,6,12,24,48,72]))['churn'].mean().round(3)
```

## Fraud — how fraudulent CDRs differ

```python
print('Fraud rate: %.2f%%' % (100 * cdrs['is_fraud'].mean()))
cdrs.groupby('is_fraud')[['duration_sec','cost','calls_last_hour','unique_numbers_called']].mean().round(1)
```

## Network — failure vs healthy KPIs

```python
print('Failure rate: %.2f%%' % (100 * network['is_failure'].mean()))
network.groupby('is_failure')[['cpu_utilization','latency_ms','packet_loss_pct','throughput_mbps','temperature_c']].mean().round(1)
```
